# Accounts Receivable

```python
import matplotlib.pyplot as plt

# List of amounts owed by customers
customers = ['Customer A', 'Customer B', 'Customer C', 'Customer D']
amounts_owed = [1000, 2000, 1500, 2500]  # Example amounts in dollars

# Calculating total accounts receivable
total_accounts_receivable = sum(amounts_owed)

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(customers, amounts_owed, color='skyblue')

# Add titles and labels
plt.title('Amounts Owed by Customers', fontsize=16)
plt.xlabel('Customers', fontsize=12)
plt.ylabel('Amount Owed (USD)', fontsize=12)
plt.axhline(y=total_accounts_receivable, color='red', linestyle='--', label=f'Total: ${total_accounts_receivable}')
plt.legend()

# Add annotations for each bar
for i, amount in enumerate(amounts_owed):
    plt.text(i, amount + 50, f"${amount}", ha='center', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

# Accounts Payable

```python
import matplotlib.pyplot as plt

# List of amounts owed to suppliers
suppliers = ['Supplier X', 'Supplier Y', 'Supplier Z', 'Supplier W']
amounts_payable = [3000, 4500, 1200, 800]  # Example amounts in dollars

# Calculating total accounts payable
total_accounts_payable = sum(amounts_payable)

# Print total accounts payable
print(f"Total Accounts Payable: ${total_accounts_payable}")

# Create the bar chart for accounts payable
plt.figure(figsize=(8, 6))
plt.bar(suppliers, amounts_payable, color='lightcoral')

# Add titles and labels
plt.title('Amounts Payable to Suppliers', fontsize=16)
plt.xlabel('Suppliers', fontsize=12)
plt.ylabel('Amount Payable (USD)', fontsize=12)
plt.axhline(y=total_accounts_payable, color='blue', linestyle='--', label=f'Total: ${total_accounts_payable}')
plt.legend()

# Add annotations for each bar
for i, amount in enumerate(amounts_payable):
    plt.text(i, amount + 100, f"${amount}", ha='center', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

# Revenue

```python
# Total sales and total discounts or returns
total_sales = 50000  # Total sales amount in dollars
total_discounts = 2000  # Total discounts or returns in dollars

# Calculating total revenue
total_revenue = total_sales - total_discounts

# Displaying the total revenue
print(f"Total Revenue: ${total_revenue}")

# Visualization
categories = ['Total Sales', 'Total Discounts', 'Total Revenue']
values = [total_sales, total_discounts, total_revenue]
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=['green', 'red', 'blue'])

# Add titles and labels
plt.title('Sales, Discounts, and Revenue', fontsize=16)
plt.ylabel('Amount (USD)', fontsize=12)

# Add annotations for each bar
for i, value in enumerate(values):
    plt.text(i, value + 500, f"${value}", ha='center', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

# Liabilities

```python
# List of liabilities
liabilities = {
    'mortgage': 200000,
    'loan': 50000,
    'credit_card': 10000
}

# Calculating total liabilities
total_liabilities = sum(liabilities.values())
print(f"Total Liabilities: ${total_liabilities}")

# Visualization
liability_types = list(liabilities.keys())
liability_amounts = list(liabilities.values())
plt.figure(figsize=(8, 6))
plt.bar(liability_types, liability_amounts, color='purple')

# Add titles and labels
plt.title('Breakdown of Liabilities', fontsize=16)
plt.xlabel('Liability Types', fontsize=12)
plt.ylabel('Amount (USD)', fontsize=12)
plt.axhline(y=total_liabilities, color='orange', linestyle='--', label=f'Total: ${total_liabilities}')
plt.legend()

# Add annotations for each bar
for i, amount in enumerate(liability_amounts):
    plt.text(i, amount + 5000, f"${amount}", ha='center', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

# Equity

```python
total_assets=500000
total_liabilities=300000
equity=200000

# Data for stacked bar chart
categories = ['Financial Overview']
asset_component = [total_assets]
liability_component = [total_liabilities]
equity_component = [equity]

# Create stacked bar chart
plt.figure(figsize=(8, 6))
plt.bar(categories, asset_component, color='green', label='Total Assets')
plt.bar(categories, liability_component, bottom=asset_component, color='red', label='Total Liabilities')
plt.bar(categories, equity_component, bottom=[liabilities + total_liabilities for liabilities in [0]], color='blue', label='Equity')

# Adding titles and labels
plt.title('Stacked Bar Chart: Assets, Liabilities, and Equity', fontsize=16)
plt.ylabel('Amount (USD)', fontsize=12)

# Adding legend
plt.legend()

# Show the chart
plt.tight_layout()
plt.show()
```

# Accruals (or Accrued Expensens)

```python
import pandas as pd
from datetime import datetime

# Example dataset of accrued expenses
accrued_expenses = [
    {'date': '2025-01-05', 'amount': 500},
    {'date': '2025-01-20', 'amount': 800},
    {'date': '2025-02-10', 'amount': 1200},
    {'date': '2025-02-25', 'amount': 700},
    {'date': '2025-03-15', 'amount': 600},
]

# Convert dataset to DataFrame
df = pd.DataFrame(accrued_expenses)

# Convert date to datetime and extract the month
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.strftime('%Y-%m')  # Extract year-month

# Group by month and sum the amounts
monthly_expenses = df.groupby('month')['amount'].sum().reset_index()

# Rename columns for clarity
monthly_expenses.columns = ['Month', 'Total Accrued Expenses']

import ace_tools as tools; tools.display_dataframe_to_user(name="Monthly Accrued Expenses Summary", dataframe=monthly_expenses)

# Visualize the monthly accrued expenses
plt.figure(figsize=(8, 6))
plt.bar(monthly_expenses['Month'], monthly_expenses['Total Accrued Expenses'], color='purple')

# Adding titles and labels
plt.title('Monthly Accrued Expenses', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Accrued Expenses (USD)', fontsize=12)

# Add annotations for each bar
for i, value in enumerate(monthly_expenses['Total Accrued Expenses']):
    plt.text(i, value + 50, f"${value}", ha='center', fontsize=10)

# Show the chart
plt.tight_layout()
plt.show()
```
