# Moving Average

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Step 1: Create a dataset of historical sales
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sales = [500, 520, 480, 510, 495, 530, 520, 490, 500, 515, 505, 520]  # Example sales data

# Step 2: Calculate a 3-month moving average
sales_series = pd.Series(sales)
moving_average = sales_series.rolling(window=3).mean()

# Step 3: Forecast the next month's sales using the last 3-month moving average
forecast = moving_average.iloc[-1]

# Step 4: Add the forecast to the dataset
months.append("Next Month")
sales_with_forecast = sales + [forecast]

# Step 5: Plot the data
plt.figure(figsize=(10, 6))
plt.plot(months[:-1], sales, label="Historical Sales", marker="o")
plt.plot(months[-2:], [sales[-1], forecast], label="Forecast", linestyle="--", color="orange", marker="o")

# Step 6: Highlight forecasted point
plt.scatter(months[-1], forecast, color="red", label=f"Forecast: {forecast:.2f}")

# Step 7: Add labels and legend
plt.title("Sales Forecasting Using Moving Average")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```

# Cash Ratio

```python
import matplotlib.pyplot as plt

# Define current assets and liabilities
current_assets = {
    'cash': 10000,
    'accounts_receivable': 30000,
    'inventory': 50000,
    'marketable_securities': 20000
}

current_liabilities = 75000  # Example current liabilities in dollars

# Calculate Current Ratio
current_ratio = sum(current_assets.values()) / current_liabilities

# Calculate Quick Ratio (excluding inventory)
quick_ratio = (
    current_assets['cash'] +
    current_assets['accounts_receivable'] +
    current_assets['marketable_securities']
) / current_liabilities

# Calculate Cash Ratio (only cash and marketable securities)
cash_ratio = (
    current_assets['cash'] +
    current_assets['marketable_securities']
) / current_liabilities

# Prepare data for visualization
ratios = {
    'Current Ratio': current_ratio,
    'Quick Ratio': quick_ratio,
    'Cash Ratio': cash_ratio
}

# Create a bar chart
fig, ax = plt.subplots(figsize=(8, 6))
bars = ax.bar(ratios.keys(), ratios.values(), color=['skyblue', 'orange', 'green'])

# Add a horizontal line at ratio = 1
ax.axhline(1, color='red', linestyle='--', linewidth=1, label='Benchmark (1.0)')

# Annotate bars with ratio values
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 5),  # Offset text above the bar
                textcoords="offset points",
                ha='center', va='bottom')

# Set chart title and labels
ax.set_title('Liquidity Ratios Visualization')
ax.set_ylabel('Ratio Value')
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

# Working Capital

```python
# Define current assets and liabilities for the example
current_assets = {
    'cash': 10000,
    'accounts_receivable': 30000,
    'inventory': 50000,
    'marketable_securities': 20000
}
current_liabilities = 75000  # Current liabilities in dollars

# Calculate Working Capital
total_current_assets = sum(current_assets.values())
working_capital = total_current_assets - current_liabilities

# Visualize Working Capital
labels = ['Total Current Assets', 'Current Liabilities', 'Working Capital']
values = [total_current_assets, current_liabilities, working_capital]
colors = ['skyblue', 'orange', 'green']

plt.figure(figsize=(8, 6))
bars = plt.bar(labels, values, color=colors)

# Annotate bars with values
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'${height:,}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 5),  # Offset text above the bar
                 textcoords="offset points",
                 ha='center', va='bottom')

# Add title and labels
plt.title("Working Capital Visualization")
plt.ylabel("Amount in USD")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

# Inventory Turnover Ratio

```python
import matplotlib.pyplot as plt

# Data for the example
cogs = 300000  # Cost of Goods Sold
average_inventory = 100000  # Average Inventory
inventory_turnover_ratio = cogs / average_inventory  # Calculate the ratio

# Data for the bar chart
categories = ['COGS', 'Average Inventory']
values = [cogs, average_inventory]

# Create the bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(categories, values, color=['skyblue', 'lightgreen'])

# Annotate the bars with their values
for bar, value in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 5000, f"${value:,}", ha='center', va='bottom')

# Add title and labels
plt.title('Inventory Turnover Analysis', fontsize=16)
plt.ylabel('Value (in $)', fontsize=12)
plt.xlabel('Categories', fontsize=12)
plt.ylim(0, max(values) + 50000)

# Annotate the turnover ratio
plt.text(0.5, max(values) - 20000, f'Inventory Turnover Ratio = {inventory_turnover_ratio:.1f}',
         fontsize=12, color='darkred', ha='center', va='top')

# Show the plot
plt.tight_layout()
plt.show()
```
