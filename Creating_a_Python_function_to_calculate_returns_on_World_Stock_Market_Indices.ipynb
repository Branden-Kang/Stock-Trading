```python
import yfinance as yf
```

```python
# creating a list of stock indices 
tickers = ['^GSPC', '^DJI', '^IXIC','^RUT', '^FTSE', '^GDAXI', '^STOXX50E', '^FCHI', '^AXJO','^N225', '^HSI', '^NSEI', '000001.SS']
#initiating the function
def get_data(symbol, start):
    # downloading data, just the 'Close' column
    data = yf.download(tickers, start=start)['Close']
    # creating a dataframe with values from the close day
    frame = frame['Close'] = data.iloc[-1].to_frame()
    # getting values for each symbol from previous day
    frame['Previous Day'] = data.iloc[-2]
    # calculating the day-to-day return
    frame['% Change'] = ((frame['Close'] - frame["Previous Day"])/frame['Previous Day'])*100
    #resetting the index, the symbols are not index column anymore
    frame = frame.reset_index()
    #renaming the symbols column to 'Symbol'
    frame.rename(columns = {'index': "Symbol"}, inplace=True)
    #creating a list of names for stock indices
    index_names = ['Shanghai Composite', 'S&P/ASX 200', 'Dow Jones', 'CAC40', 'FTSE 100', 'DAX','S&P 500', 'Hang Seng', 'NASDAQ', 'Nikkei 225','Nifty 50', ' Russell 2000', ' EURO STOXX 50']
    #creating a column with stock names
    frame['Index Name'] = index_names
    #creating a column that indicates whether the index went up or down from previous day
    frame['Direction'] = frame['% Change'].apply(lambda x: 'Up' if x >= 0 else 'Down')
    #getting values from row 0 to calculate YTD return
    frame['Day 1 of Year'] = data.iloc[0].values
    #calculating YTD return
    frame['YTD %'] = ((frame['Close'] - frame["Day 1 of Year"])/frame['Day 1 of Year'])*100
    #creating a column that indicates whether the YTD is negative or positive
    frame['Overall YTD Return'] = frame['YTD %'].apply(lambda x: 'Positive' if x >= 0 else 'Negative')
    #dropping first column used to create 'frame' dataframe
    frame.drop(frame.columns[1], axis=1, inplace=True)
    #reorganizing the dataframe
    frame = frame[['Symbol', 'Index Name', 'Close','Previous Day', '% Change', 'Direction','YTD %', 'Overall YTD Return']]
    #closing the function
    return frame
#now the 'tickers' list can be plugged into the function 
data = get_data(tickers, "2022-01-04")
```
